```python
df = df[['Date','Amount', 'Category', 'ship-state', 'Size', 'Qty', 'Status', 'Courier Status']]
```

```python
df['Status'] = df['Status'].str.replace('Shipped - Delivered to Buyer','Shipped')
df['Status'] = df['Status'].str.replace('Shipped - Returned to Seller','Returned')
df['Status'] = df['Status'].str.replace('Shipped - Picked Up','Shipped')
df['Status'] = df['Status'].str.replace('Shipped - Returning to Seller','Returning')
df['Status'] = df['Status'].str.replace('Shipped - Out for Delivery','Out for Delivery')
df['Status'] = df['Status'].str.replace('Shipped - Rejected by Buyer','Rejected')
df['Status'] = df['Status'].str.replace('Shipped - Damaged','Damaged')
df['Status'] = df['Status'].str.replace('Shipping','Pending')
df['Status'] = df['Status'].str.replace('Shipped - Lost in Transit','Lost')
df['Status'] = df['Status'].str.replace('Pending - Waiting for Pick Up','Pending')
```

```python
df
```

```python
df = df[(df['Status'] == 'Shipped') & (df['Courier Status'] == 'Shipped')]
```

```python
df
```

```python
df.info()
```

```python
df = df.reset_index(drop=True)
```

```python
df
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%m-%d-%y')
```

```python
df
```

```python
df = df.rename(columns={'ship-state':'Ship_State',
                  "Courier Status":'Courier_Status'})
```

```python
df['Day'] = df['Date'].dt.day_name()
```

```python
df['Month'] = df['Date'].dt.strftime('%b')
```
